## Board generator for the 6 player field

            
See the 4 player version for more details

```python
import json
from step_generator import Stepper
```

```python
# Generate the steps for 6 players

height = 800
width = 800
hw_ratio = height/width

# let u be the distance between the first and the last node in a line of 5
# 1u is 21% of the board width
u = 15/100*width

allsteps = []
outer = []    # circles around homes and houses and the base step
homes = []
houses = []


def do4(f):
    for i in range(4):
        allsteps.append(step.get_step())
        f()
        
# start at top right corner, the initial values are computed from the middle
# of the board
step = Stepper(width/2 + 1.2*u*hw_ratio, height/2 - 2.9*u, u)

base1 = step.get_coords()
outer.append(step.get_outer(0))

do4(step.sw)
do4(step.w)
do4(step.nw)
do4(step.sw)

base2 = step.get_coords()
outer.append(step.get_outer(1))

do4(step.se)
do4(step.s)
do4(step.w)
do4(step.s)

base3 = step.get_coords()
outer.append(step.get_outer(2))

do4(step.e)
do4(step.s)
do4(step.sw)
do4(step.se)

base4 = step.get_coords()
outer.append(step.get_outer(3))

do4(step.ne)
do4(step.e)
do4(step.se)
do4(step.ne)

base5 = step.get_coords()
outer.append(step.get_outer(4))

do4(step.nw)
do4(step.n)
do4(step.e)
do4(step.n)

base6 = step.get_coords()
outer.append(step.get_outer(5))

do4(step.w)
do4(step.n)
do4(step.ne)
do4(step.nw)

# Generate the homes
step = Stepper(0, 0, u)

def do4(f, color_id):
    f(scale=1.4)
    homes.append(step.get_home(color_id))
    outer.append(step.get_outer(color_id))
    for i in range(3):
        f()
        homes.append(step.get_home(color_id))
        outer.append(step.get_outer(color_id))

# Straight homes for center-left and center-right player
homebuilding = [(base1, step.w, 0), (base2, step.s, 1), (base3, step.s, 2),
                (base4, step.e, 3), (base5, step.n, 4), (base6, step.n, 5)]

for base, direction, color_id in homebuilding:
    step.set_coords(*base)
    do4(direction, color_id)

# Generate the houses
step = Stepper(0, 0, u)


def draw_house(f1, f2, color_id):
    f1(scale=1.4)
    houses.append(step.get_house(color_id))
    outer.append(step.get_outer(color_id))
    f1()
    houses.append(step.get_house(color_id))
    outer.append(step.get_outer(color_id))
    for i in range(2):
        f2()
        houses.append(step.get_house(color_id))
        outer.append(step.get_outer(color_id))
        

housebuilding = [(base1, step.s, step.sw, 0), (base2, step.e, step.se, 1), (base3, step.ne, step.e, 2),
                 (base4, step.n, step.ne, 3), (base5, step.w, step.nw, 4), (base6, step.sw, step.w, 5)]

for base, dir1, dir2, color_id in housebuilding:
    step.set_coords(*base)
    draw_house(dir1, dir2, color_id)
```

```python
# data dictionary that will be saved as a json file
data = {'steps': allsteps, 'outer': outer, 'homes': homes, 'houses': houses}

with open('../src/components/board/boarddata6.json', 'w') as fp:
    json.dump(data, fp)
```
